```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
np.random.seed(240)
img = plt.imread("to the moon.jpg")
offsetArr = np.random.randint(low=0, high=255, size=(img.shape))
```

```python
print(img.shape, offsetArr.shape)
```

```python
def deviate(img_org, val):
    return (img_org + val) % 256
```

```python
offset = np.vectorize(deviate)
img_flat = img.reshape(-1)
offset_flat = offsetArr.reshape(-1)
img_encrypted = offset(img_flat, offset_flat)
```

```python
plt.imshow(img_encrypted.reshape(img.shape))
```

```python
def restore(encrypted, offset):
    return 256 + encrypted - offset
```

```python
recover = np.vectorize(restore)
img_restored = recover(img_encrypted, offset_flat).reshape(img.shape)
```
