```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
np.random.seed(240)
img = plt.imread("to the moon.jpg")
offsetArr = np.random.randint(low=0, high=255, size=img.shape[0])
```

```python
print(img.shape, offsetArr.shape)
```

```python
def deviate(matrix, val):
    return matrix + val
```

```python
offset = np.vectorize(deviate)
for i in range(img.shape[0]):
    img[i] = offset(img[i], offsetArr[i])
```

```python
plt.imshow(img)
```

```python
def restore(matrix, val):
    return matrix - val
```

```python
np.random.seed(240)
offsetArr = np.random.randint(low=0, high=255, size=img.shape[0])
recover = np.vectorize(restore)
for i in range(img.shape[0]):
    img[i] = recover(img[i], offsetArr[i])
```

```python
plt.imshow(img)
```
